```python
import os
from bs4 import BeautifulSoup
import re


for i in range(1,5):
	directory = "articles_by_category/category{}".format(i)
	#print(directory)
	count = 0
	for file in os.listdir(directory):
		#print(file)
		content = """"""
		read_addr = "{}/{}".format(directory, file)

		with open(read_addr, encoding = 'ISO-8859-1') as response:
			soup = BeautifulSoup(response, 'html.parser')

		find_results = soup.findAll('p', attrs={'class': 'story-body-text story-content'})
		for paragraph in find_results:
			content += paragraph.get_text()

		if (content):
			count += 1
			write_addr = "{}/data_{}.txt".format(directory, count)
			#content = content.replace("â", "")
			with open(write_addr, 'a+', encoding = 'ISO-8859-1') as write_content:
				write_content.write(content)

```

```python
import pandas as pd
import numpy as np 
import os 
import re
from collections import deque

df = pd.DataFrame()
database = deque()
c = 0
for i in range(1,5):
	directory = "articles_by_category/category{}".format(i)
	if i == 1:
		category = 'sports'
	elif i == 2:
		category = 'politics'
	elif i == 3:
		category = 'business'
	else :
		category = 'education'

	for file in os.listdir(directory):
		aux = list()
		if (file[-4:] == '.txt'):
			c += 1
			read_addr = "{}/{}".format(directory, file)
			with open(read_addr, 'r+') as content:
				database.append([category, content.read()])

array = np.array(database)
df = pd.DataFrame(array)
print(df)
df.to_csv('data.csv', index = False, header = False)


```
