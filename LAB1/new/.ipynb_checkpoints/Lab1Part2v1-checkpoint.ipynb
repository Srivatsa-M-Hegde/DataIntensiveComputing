```r
#plot for national summary 
library(maps)
library(ggplot2)
influenza_heat_map = read.csv('./heat_map.csv')
states = map_data('state')
influenza_heat_map$region = tolower(influenza_heat_map$STATENAME)
influenza_map = merge(states, influenza_heat_map, by = "region")
c_array = c("green","firebrick4","olivedrab3","yellow","goldenrod3","goldenrod4","orange","red", "firebrick3")
ggplot(influenza_map, aes(x=long,y=lat,group=group)) + geom_polygon(aes(fill=ACTIVITY.LEVEL)) + # using polygon fill for each states depending on ACTIVITY.LEVEL
scale_fill_manual(values = c_array) + 
coord_map() +
ggtitle("Influenza Spread for the Jan 27 Week") 
```

```r
mortality = read.csv('./mortality.csv')

# for deaths due to P and I
x = c(1:432)
y = c(mortality[,3])
plot(x, y, col="red", type = "l", xlab = "MMWR", ylab = "Death Percentage due to Pnemonia & Influenza", main="Pneumonia and Influenza mortality for Jan 27 Week")
par(new = TRUE)

# for expected mortality
y <- c(mortality[,4])
plot(x,y,col = "blue", type = "l", xaxt = 'n', yaxt = 'n', ann = FALSE)

# for actual mortality 
par(new = TRUE)
y <- c(mortality[,5])
plot(x,y, col = "green",type = "l", xaxt = 'n', yaxt = 'n', ann = FALSE)
```

```r
library(htmltab) 

url = "https://www.cdc.gov/flu/weekly/weeklyarchives2017-2018/data/whoAllregt_phl04.html" 
test = htmltab(doc = url, which = "//th[text() = 'Week']/ancestor::table") 


data = data.frame(test[9], test[8], test[7], test[2], test[4], test[3], test[6])
mat = as.matrix(data)  
x = test[1] 
barplot(t(mat),ylim = c(0,4000),names.arg=t(x), xlab = "Week", ylab = "Tested Positive",col = rainbow(7))


legend("topleft", inset = 0.02, legend = c("B (Yamagata Lineage)", "B (Victoria Lineage)", "B (lineage not found)", "H32NV", "A(H3N2)", "A(H1N1)",  "A(subtyping not performed)"),
       col=rainbow(7), lty=1:7, cex=0.8)
```

```r
fun <- function (flag, infl_list) { 
    x = infl_list[flag==TRUE,4]
    x = sub("%","",x) 
    x = as.numeric(x)  
    x = x[x!=0] #remove 0  
    x = x[!is.na(x)] # remove NA  
    return (x)
}

infl_list <- read.csv("./influenza_subtype.csv")  

flag1 = infl_list[1] == 'B/Victoria'  
flag2 =  infl_list[1] == 'B/Yamagata' 
flag3 = infl_list[1] == 'H1pdm09'
flag4 = infl_list[1] == 'H3'


par(mfrow = c(2,2)) #
x1 = fun(flag1,infl_list)  
x2 = fun(flag2,infl_list)
x3 = fun(flag3,infl_list)
x4 = fun(flag4,infl_list)

s1 = paste(infl_list[9,2], infl_list[9,3] , infl_list[9,4]) 
s2 = paste(infl_list[10,2] , infl_list[10,3] , infl_list[10,4])
s3 = paste(infl_list[11,2], infl_list[11,3] , infl_list[11,4])
pie(x4, col = rainbow(3), border = "white", main = "Influenza A(H3N2)",
labels = c(s1,s2,s3)
)


s1 = paste(infl_list[6,2], infl_list[6,3] , infl_list[6,4])
pie(x3, col = rainbow(3), border = "white", main = "Influenza A(H1N1)pdm09",
label = s1)


s1 = paste(infl_list[1,2], infl_list[1,3] , infl_list[1,4])
s2 = paste(infl_list[2,2], infl_list[2,3] , infl_list[2,4])

pie(x1, col = rainbow(3), border = "white", main = "Influenza B Victoria",
label = c(s1,s2))


s1 = paste(infl_list[3,2], infl_list[3,3] , infl_list[3,4])
pie(x2, col = rainbow(3), border = "white", main = "Influenza B Yamagata",
label = s1)





```
